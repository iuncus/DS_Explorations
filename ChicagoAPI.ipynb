```python
import requests
import pandas as pd
import os
import urllib.request
from urllib.request import urlopen
from urllib.error import URLError, HTTPError
```

slap the search term onto the back of the api url

```python

url = "https://api.artic.edu/api/v1/artworks"
search_term = "Academic"
endpoint = "/search?q=" + search_term 
url = url+endpoint
```

just normal response stuff, get the status code, grab the content json

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
response = requests.get(url, headers = headers)
response.text
response.status_code
```

```python
term_json = response.json()
term_json 
```

put the content of each link the search turned up into a dataframe 

```python
results = []
for i in term_json['data']:
    link = i["api_link"]
    response = requests.get(link).json()
    results.append(response)
# api_links = [item['api_link'] for item in term_json['data']]
# print(api_links)
df = pd.json_normalize(results)
```
